```python
corpus = [
    "the girl played with the toy",
    "the girl bought a toy",
    "the girl bought a chocolate",
    "the boy ate the chocolate"
]
```

```python
tokens = []
for sentence in corpus:
    sentence = sentence.lower()
    words = sentence.split()
    for w in words:
      tokens .append(w)
```

```python
unigrams = {}
for word in tokens:
  if word in unigrams:
    unigrams[word] += 1
  else:
    unigrams[word] = 1
```

```python
bigrams = {}
for i in range(len(tokens)-1):
  pair = (tokens[i], tokens[i+1])

  if pair in bigrams:
    bigrams[pair] += 1
  else:
    bigrams[pair] = 1
```

```python
bigram_prob = {}

for pair in bigrams:
  w1 , w2 = pair
  bigram_prob[pair] = bigrams[pair]/unigrams[w1]
```

```python
def predict_next_word(word):
  max_prob = 0
  next_word = None

  for (w1 , w2),prob in bigram_prob.items():
    if w1 == word:
      if prob > max_prob:
        max_prob = prob
        next_word = w2
  return next_word
```

```python
input_word = 'ate'
prediction = predict_next_word(input_word)
print(f"The next word after '{input_word}' is '{prediction}'.")
```
