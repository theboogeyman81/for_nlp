```python
clean_text = ""
for ch in text:
  if not ch.isdigit():
    clean_text += ch
text = clean_text
```

```python
tokens = text.split()
```

```python
stopwords = ['i' , 'am' , 'is' , 'in' , 'at' , 'or' , "me"]
filtered_tokens = []
for w in tokens:
  if w not in stopwords:
    filtered_tokens.append(w)

print("Final Tokens: ")
print(filtered_tokens)
```
